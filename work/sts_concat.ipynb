```python
import pandas as pd
import glob
import os
```

```python
pd.set_option('display.max_columns', 500)
```

```python
sts_dir = '/Users/am60/Downloads/STS manifests/Uploaded to STS'
```

```python
sts_fns = glob.glob(sts_dir + '/*.xlsx')
sts_dfs = []
for sts_fn in sts_fns:
    try:
        sdf = pd.read_excel(sts_fn, sheet_name='Metadata Entry')
        sdf['FILENAME'] = sts_fn.split('/')[-1]
        sts_dfs.append(sdf)
    except:
        print(f'{sts_fn} is not an STS manifest, skipping')
    
len(sts_dfs)
```

```python
sts_df = pd.concat(sts_dfs)
sts_df.shape
```
